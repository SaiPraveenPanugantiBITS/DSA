# Algorithm benchmarking
#### *Comparison of Bubble sort & Selection sort*

* Helps you to test algorithm performance
* Compare algorithms
* Measure the executed time
* Count the number of steps taken by each algorithm (Comparison & Swap operations)
* Growth of algorithm's time complexity function
* Effect of array size and best/worst case on algorithm's performance.



```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def selection_sort(A):
  start = time.time()
  c = 0
  s = 0
  for i in range(len(A)):
    min_= i
    #print("i =", i, ", A[i] =", A[i], "Min is ", A[min_])
    for j in range(i+1, len(A)):
      #print("j =", j, ", A[j] =", A[j])
      #print("Comparing",i,j)
      c += 1
      if A[min_] > A[j]:
        min_ = j
        #print("Min is ", A[min_])
    if(A[i]!=A[min_]):
      s += 1
      #print("Swapping",A[i],A[min_])
      A[i], A[min_] = A[min_], A[i]
      #print(A)
  #print("Comparisons = ",c,"| Swapping =",s, "| Total = ",c+s)
  end = time.time()
  t = end-start
  t = c+s # Comment this line if you want to measure time taken instead of no. of steps
  return(t)
```

```python
def bubbleSort(A):
  start = time.time()
  c = 0
  s = 0
  n = len(A)
  for i in range(n):
    for j in range(i+1, n):
      # print("Comparing",i,j)
      c+=1
      if A[i] > A[j]:
        # print("Swapping",A[i],A[j])
        s+=1
        # print(A)
        A[i], A[j] = A[j], A[i]
        # print(A)
  #print("Comparisons = ",c,"| Swapping =",s, "| Total = ",c+s)
  end = time.time()
  t = end-start
  t = c+s # Comment this line if you want to measure time taken instead of no. of steps
  return(t)
```

## Question no. 1

```python
n = 30
sel_best = [None] * n
sel_worst = [None] * n
bub_best = [None] * n
bub_worst = [None] * n

for k in range(0,n):
  Asb = np.arange(1, k+1) # Best case Array
  Asw = np.arange(k, 0, -1) # Worst case Array
  Abb = np.arange(1, k+1) # Best case Array
  Abw = np.arange(k, 0, -1) # Worst case Array
  sel_best[k] = selection_sort(Asb)
  sel_worst[k] = selection_sort(Asw)
  bub_best[k] = bubbleSort(Abb)
  bub_worst[k] = bubbleSort(Abw)

# print(sel_best)
# print(sel_worst)
# print(bub_best)
# print(bub_worst)
```

```python
plt.figure(figsize=(14,8))
plt.plot(sel_best, linestyle = 'dashed', label = 'Best case growth for Selection/Bubble sort') # Plotting Selection sort best case growth
plt.plot(sel_worst, linestyle = 'solid', label = 'Worst case growth for Selection sort') # Plotting Selection sort worst case growth
plt.plot(bub_best, linestyle = 'dotted') # Plotting Bubble sort best case growth
plt.plot(bub_worst, linestyle = 'solid', label = 'Worst case growth for Bubble sort') # Plotting Bubble sort worst case growth
plt.legend()
plt.show()
```

```python
k = 454
BAB = np.arange(1,k+1) # Best case array
BAW = np.arange(k, 0, -1) # Worst case array
bab = bubbleSort(BAB)
baw = bubbleSort(BAW)
print("Number of total steps for Bubble sort best case =",bab) # If n = 454 for best case it is 102831
print("Number of total steps for Bubble sort worst case =",baw) # If n = 454 for worst case it is 205662
```

```python
k = 454
SAB = np.arange(1, k+1) # Best case array
SAW = np.arange(k, 0, -1) # Worst case array
sab = selection_sort(SAB)
saw = selection_sort(SAW)
print("Number of total steps for Selection sort best case =",sab) # If n = 454 for best case it is 102831
print("Number of total steps for Selection sort worst case =",saw) # If n = 454 for worst case it is 103058
```
